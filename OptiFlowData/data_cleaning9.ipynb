```python
def aggregate_to_hourly(df):
  df['observation_time'] = pd.to_datetime(df['observation_time'])

  df['hour'] = df['observation_time'].dt.floor('H')

  df_hourly = df.groupby(['reservoir_id', 'hour']).mean(numeric_only=True)

  df_hourly.reset_index(inplace=True)

  hourly_data = hourly_data.rename(columns={
    'hour': 'observation_time', 
  })

  # 원하는 칼럼 순서 정의
  new_order = ['observation_time', 'reservoir_id', 'input', 'output', 'height']

  # 순서 변경 적용
  hourly_data = hourly_data[new_order]

  return df_hourly
```

```python
hourly_data = aggregate_to_hourly(data)
hourly_data.head()
```
