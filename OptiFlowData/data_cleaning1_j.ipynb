```python
from preamble import *
```

```python
data9 = pd.DataFrame(pd.read_csv('data/9.csv', header=None))
data10 = pd.DataFrame(pd.read_csv('data/10.csv', header=None))
data11 = pd.DataFrame(pd.read_csv('data/11.csv', header=None))
data12 = pd.DataFrame(pd.read_csv('data/12.csv', header=None))
data13 = pd.DataFrame(pd.read_csv('data/13.csv', header=None))
```

```python
input_flow = data9.set_index(1)[2]
output_flow_1 = data10.set_index(1)[2]
output_flow_2 = data11.set_index(1)[2]
height_new = data12.set_index(1)[2]
height_old = data13.set_index(1)[2]
input_flow.head()
# print(type(input_flow))
```

```python
j_data = pd.DataFrame({'input_flow':input_flow, 'output_flow_1':output_flow_1, 'output_flow_2':output_flow_2, 'height_new':height_new, 'height_old':height_old})
j_data.index.name = 'Date'
j_data.index = pd.to_datetime(j_data.index)
j_data.head()
```

```python
df = pd.DataFrame(pd.read_csv('./data/weather_2023.csv'))
df2 = pd.DataFrame(pd.read_csv('./data/weather_2024.csv'))
weather_2023 = df.iloc[:, 2:]
weather_2024 = df2.iloc[:, 2:]
weather_2023 = weather_2023.set_index(weather_2023.columns[0]) 
weather_2024 = weather_2024.set_index(weather_2024.columns[0])
weather_2023.fillna(0, inplace=True)
weather_2024.fillna(0, inplace=True)
weather_2023.head(), weather_2024.head()
```

```python
weather = pd.concat([weather_2023, weather_2024], axis=0)
weather.head()
# weather.tail()
```

```python
weather.index = pd.to_datetime(weather.index)
weather.index.name = 'Date'
weather_min = weather.resample('min').ffill().bfill()
```

```python
weather_min.rename(columns={'기온(°C)': 'temperature', '강수량(mm)': 'precipitation', '풍속(m/s)': 'wind_speed', '습도(%)': 'humidity', '현지기압(hPa)':'atmospheric_pressure', '적설(cm)':'snow'}, inplace=True)
weather_min.head()
```

```python
data = pd.concat([j_data, weather_min], axis=1, join='inner')
data.head()
```

```python
output_flow = j_data['output_flow_1'] + j_data['output_flow_2']
# label = pd.DataFrame({
#     'output_flow': output_flow
# }, index=data.index)
# label.head()
```

```python
data_dropped = data.drop(['output_flow_1', 'output_flow_2'], axis=1)
data_dropped['output_flow'] = output_flow
```

```python
data_dropped = data_dropped.astype(float)
data_dropped.shape
```

```python
data_na = data_dropped.dropna()
data_na.shape
```

```python
data_na.to_csv(path_or_buf='data/j_data.csv')
# label.to_csv(path_or_buf='data/j_label.csv')
```
